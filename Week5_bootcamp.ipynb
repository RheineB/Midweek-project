# Hypothesis testing

##### In this section of the lab, we will test whether the monthly savings or the extra money that people have at the end of the year is significantly different from 3271.40 that people in Navarra save.

```python
import scipy.stats as st
from scipy.stats import ttest_1samp
```

```python
savings = data['savings'].dropna()
```

```python
#Two-sided test:
    # Null hypothesis or H0: savings = 3271.40/year
    # Alternative hyp or H1: savings != 3271.40/year

stat, pval = ttest_1samp(savings, 3271.40)

print('stat is  ', stat)
print('pvalue for the two-tailed test is  ', pval)
print('pvalue for the one-tailed test is  ', pval/2)
```

##### based on the hypothesis, we reject the null hypothesis
